```python
import numpy as np
from numpy import sin, cos, tan, atan, cosh, sinh, tanh, abs, linspace, min, max, argmin, argmax, pi, mean, exp, sqrt, zeros, ones, nan
import scipy
import matplotlib.pyplot as plt

```

```python


phi = pi / 6 # 30 degrees
L = 8e3
H = 7.12

h_x_dx = lambda x: 1
h_x_dx = 1
```

```python
# alpha_s = tan(phi) / 

c1 = tan(phi)
c2 = cos(atan(H/L * h_x_dx)) # ~ 1
c3 = H / L * h_x_dx # very small

k2 = c2 * c1 
k3 = c2 * c3 # very small


print(c1)
print(k2)
print(k3)
```

```python
u = linspace(-2, 2, 1000)

alpha = c1 / (k2 + k3 * np.sign(u))

plt.plot(u, alpha)
plt.show()

```
