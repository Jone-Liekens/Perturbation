```python

from sympy import symbols, Function, diff, expand, collect, pprint, latex, \
    sin, cos, integrate, O, simplify, pi, init_session, init_printing
import sympy

import matplotlib

init_printing()
```

```python

# for interactive session
# init_session()

# for notebooks:
# init_printing(use_latex='mathjax')
# import IPython.display as disp

x = symbols('x')
t = symbols('t')
eps = symbols("self.epsilon")


u1_r = Function('u1r_x')(x)
u1_c = Function('u1c_x')(x)
u1_s = Function('u1s_x')(x)

u0_c = Function('u0c_x')(x)
u0_s = Function('u0s_x')(x)




u1 = u1_r + u1_c * cos(2*t) + u1_s * sin(2*t)
u0 = u0_c * cos(t) + u0_s * sin(t)
u = u0 + eps * u1


# alpha_s = 3.1 / (3.2 + 3.3 * sympy.sign(u))


# ignore terms of order eps**2
# u_comp = expand(diff(alpha_s * (u)**5, x)) + O(eps**2)
u_comp = expand(u**5) + O(eps**2)


q = integrate(u_comp, (t, 0, 2*pi))


q = diff(q, x)
q = simplify(q)
q = collect(q, eps)
# q = collect(q, u0_c)

# pprint(q)

l = str(q)
# print(len(l))
l = l.replace('Derivative(u1r_x(x), x)', 'u1r_x_dx')
l = l.replace('Derivative(u1c_x(x), x)', 'u1c_x_dx')
l = l.replace('Derivative(u1s_x(x), x)', 'u1s_x_dx')
l = l.replace('Derivative(u0c_x(x), x)', 'u0c_x_dx')
l = l.replace('Derivative(u0s_x(x), x)', 'u0s_x_dx')
l = l.replace('(x)', '')
l = l.replace(' + O(self.epsilon**2)', '')
# print(len(l))
print(l)






```
