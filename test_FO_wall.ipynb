```python
%load_ext autoreload
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm
from numpy.polynomial import chebyshev
from cycler import cycler
```

```python
from pcswe import PerturbationCSWE
```

```python
tau = 0.1

h_fx = lambda x: 0.9*((1+tau) * x + (-tau) * x**2)
h_fx_dx = lambda x: 0.9*(1 + tau + 2*(-tau)*x)
h_fx_dxx = lambda x: 0


pcswe = PerturbationCSWE()
pcswe.H = 7.12
pcswe.A = 0.72
pcswe.L = 8e3
pcswe.r = 0.3
pcswe.h0 = 0.1

pcswe.h_fx = h_fx
pcswe.h_fx_dx = h_fx_dx
pcswe.h_fx_dxx = h_fx_dxx

pcswe.tol = 1e-6
pcswe.small_number = 1e-7

pcswe.bc = pcswe.bc_wall

pcswe.set_derivative_vars()

# pcswe.solve()
# pcswe.visualize_components()

pcswe.solve()
pcswe.visualize_components()

pcswe.visualize_transport()
```

```python

n_colors = 40
colors = plt.cm.viridis(np.linspace(0, 1, n_colors-1))
custom_cycler = (cycler(color=colors[::-1]))
plt.rcParams['axes.prop_cycle'] = custom_cycler


# pcswe.lmbda = 0

timestep = 2e-2

fig, axs = plt.subplots(2, 3, figsize=(30, 20))
for i in range(40):
    pcswe.transport_step(timestep)
    if i % 2 == 0:
        # pcswe.visualize_components()
        # pcswe.visualize_transport()

        x, u5_xt, q_xt, q_x, q_x_trap, q_x_dx, q_x_dx_num, b_x, b_x_dx, h_x_dt = pcswe.transport()

        
        axs[0, 0].plot(x, pcswe.h_fx(x))#, 'brown', linewidth=4)
        # axs[0, 1].plot(x, q_x_trap, 'o')
        # axs[0, 1].plot(x, q_x_gl, 'x', ms=5)
        axs[0, 1].plot(x, q_x)#, 'k')
        axs[0, 2].plot(x, b_x)

        axs[1, 0].plot(x, h_x_dt)#, color='brown', linestyle='-')
        axs[1, 1].plot(x, q_x_dx)
        axs[1, 2].plot(x, b_x_dx)

plt.show()
```
